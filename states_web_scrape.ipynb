```python
level1 = ['RURAL', "URBAN"]
level2 = ['INTERSTATE', 'OTHER FREEWAYS AND EXPRESSWAYS','OTHER PRINCIPAL ARTERIAL', 'MINOR ARTERIAL', 'MAJOR COLLECTOR', 'MINOR COLLECTOR', "LOCAL", "TOTAL"]
files = "2009_hm80.xls 2012_hm80.xls 2014_hm80.xls 2016_hm80.xls 2018_hm80.xls 2011_hm80.xls 2013_hm80.xls 2015_hm80.xls 2017_hm80.xls".split(" ")
```

```python
print(files)
total_data = []

for file in files:
    print(file)
    year = int(file.split("_")[0])
    cols = pd.MultiIndex.from_product([level1, level2])

    path = "./miles/{}".format(file)
    data_1 = pd.read_excel(path, skiprows=13,sheet_name="A", na_values=0)    
    cols = ['state'] + cols.to_list() + [('TOTAL', "TOTAL")]
    data_1.columns = cols
    data_1['date'] = year
    data_1 = data_1.drop(index=[53, 54])
    total_data.append(data_1)
```

```python
tot_df = pd.concat(total_data)
def change(string):
    string = string.replace("2/", "")
    string = string.replace("(2)", "")
    if string == "Dist. of Columbia":
        return "District of Columbia"
    return string.strip()
tot_df['state'] = tot_df['state'].map(change)

all_states = tot_df['state'].unique()
print(all_states)
```

```python
tot_df
```

```python
for state in all_states:
    print(state)
    dta = tot_df.loc[tot_df['state'] == state]
    index = dta.columns.to_list()
    index.remove('date')
    index.insert(0, "date")
    dta = dta[index].sort_values(by="date")
    dta.to_csv("./miles/{}_miles.csv".format(state), index=False)
```
